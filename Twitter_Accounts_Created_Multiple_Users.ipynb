# DFRLab's Notebook on finding 'account created' date for 100 accounts using Twitter API. 


```python
#Installing packages for the notebook. 

from twython import Twython
#Twython is a library for providing easy way to scrape data. 
import pandas as pd
#Pandas is a library for analtyzing data. 
import csv
#CSV is the format required for storing the data in a file. 


#add your Twitter Access Keys
t = Twython(app_key='', 
    app_secret='',
    oauth_token='',
    oauth_token_secret='')

#^Please do not share your keys with anyone, as keys could be misused. You can find your keys, or make one for your application here: 
#https://apps.twitter.com/ 
```

```python

# Add your data inside the data = [], for example: data = ["John Kane", "Ralph Scott"]
data = []

output = t.lookup_user(screen_name=data) 
#Here's where we parse the data with Twitter.
accountcreated_list=[]
#This is where we encapsulate the data, showing how different accounts have different account_created dates. 

#Loop through the results
for user in output:
    accountcreated_list.append(user['created_at'])
    
#The reason why we are using a 'for' statement here, is to make the process easier, instead of putting a query in each line for different accounts, we are basically asking Twitter to give us the list of the accounts as mentioned.  
print accountcreated_list 
#The upper line prints the list. 
```

```python
#Here is where we use, Pandas to bind the data in adjacent to the accountcreated_list.
pd.DataFrame(data, accountcreated_list)
```

```python
#Here's where we save the data to a csv file. 
frame = zip(data ,accountcreated_list)
df = pd.DataFrame(frame)

#Saving the file to csv.
df.to_csv('accountscreated.csv', index=False)
```
